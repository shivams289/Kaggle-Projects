```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stats
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

# Normal Distribution

```python
#1 million sample points with mean 5 and standard deviation10
#Normal distribution is a continuos distribution
dist_norm = np.random.normal(loc = 5, scale =0.5, size = 1000000)
```

```python
#How it looks
sns.displot(dist_norm, kde=True)
print(f'Mean:{np.mean(dist_norm)}\nMedian:{np.median(dist_norm)}\nStdv:{np.std(dist_norm)}')
```

```python
#whether number of datapoints in a particular quantile is same as theoretical quantiles for normal dist
norm_plot = sm.qqplot(dist_norm,line='45',fit=True,dist=stats.norm)
```

## Bootstrapping sample means and according to central limit theorem the distribution of means of samples taken from any distribution(except Cauchy) will follow a Normal distribution

```python
#taking random samples and storing their means
sample_means = []
for i in range(1000):
    temp = np.random.choice(dist_norm, size = 100, replace = True)
    sample_means.append(temp.mean())
    
sample_means = np.array(sample_means)
```

```python
#seems to be true, with approx same mean median, here standard deviation is the standard error:standard deviation of sample means
sns.displot(sample_means, kde=True)
print(f'Mean:{np.mean(sample_means)}\nMedian:{np.median(sample_means)}\nStdv:{np.std(sample_means)}')
```

```python
## Yes: the sample means follow a normal distribution
norm_plot = sm.qqplot(sample_means,line='45',fit=True,dist=stats.norm)
```

```python
#Confidence interval for mean
np.percentile(sample_means, [2.5,97.5])
```

# Poisson Distribution

```python
#lam:Expected number of events occurring in a fixed-time interval
#Poisson Distribution is a Discrete Distribution.
#It estimates how many times an event can happen in a specified time. 
#e.g. If someone eats twice a day what is probability he will eat thrice?

dist_poisson = np.random.poisson(lam = 3, size = 1000000)
```

# Exponential Distribution

```python
#Exponential distribution is used for describing time till next event
#Scale: inverse of rate lambda
#Poisson distribution deals with number of occurences of an event in a time period whereas
#Exponential distribution deals with the time between these events.

dist_expo1 = np.random.exponential(scale = 2, size = 1000)
dist_expo2 = np.random.exponential(scale = 1, size = 1000)
dist_expo3 = np.random.exponential(scale = 0.5, size = 1000)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import math
from prettytable import PrettyTable
```

```python
df = pd.read_csv('Bid_ask_g10.csv')
df.Dates = pd.to_datetime(df['Dates'], infer_datetime_format=True)
```

```python
bid_cols = list(set(df.columns[df.columns.str.contains('BID')].values))
ask_cols = list(set(df.columns[df.columns.str.contains('ASK')].values))
bid_cols.sort()
ask_cols.sort()
```

```python
print(bid_cols)
print(ask_cols)
```

```python
for i,v in zip(bid_cols,ask_cols):
    df['spot'+i[7:]] = (df[i]+df[v])/2
```

```python
spot_cols = list(set(df.columns[df.columns.str.contains('spot')].values))
spot_cols.sort()
```

```python
for i in spot_cols:
    df['return_'+i[4:]] = (df[i]/df[i].shift(1) - 1)*100
```

```python
return_cols = list(set(df.columns[df.columns.str.contains('return')].values))
return_cols.sort()
returns = df[return_cols]
```

```python
#(ask price – bid price) / (ask price + Bid Price)

for i,v in zip(ask_cols, bid_cols):
    df['liquidity_'+i[7:]] = (df[i] - df[v])/(df[i] + df[v])
```

```python
liquid_cols = list(set(df.columns[df.columns.str.contains('liquidity')].values))
liquid_cols.sort()
liquidities = df[liquid_cols]
```

```python
#csfs:cross_sectional_factor_score
for i in liquid_cols:
    df['csfs_'+i[10:]] = 1/(1+np.exp(-(df[i] - df[liquid_cols].mean(axis=1))/df[liquid_cols].std(axis=1))) 
```

```python
csfs_cols = list(set(df.columns[df.columns.str.contains('csfs')].values))
csfs_cols.sort()
csfs = df[csfs_cols]
```

```python
#benchmark return:equally Weighted
df['bench_return'] = df[return_cols].mean(axis=1)
```

```python
#portfolio weight of currency i = csfs of i*1/9 / summation of csfs*1/9
for i in csfs_cols:
    df['pw_'+i[5:]] = df[i]/df[csfs_cols].sum(axis=1)
```

```python
csfs
```

```python
pw_cols = list(set(df.columns[df.columns.str.contains('pw')].values))
pw_cols.sort()
portfolio_weights = df[pw_cols]
```

```python
portfolio_weights
```

```python
returns.columns
```

```python
returns.columns = ['AUD', 'CAD', 'CHF', 'EUR', 'GBP', 'JPY', 'NOK', 'NZD', 'SEK']
portfolio_weights.columns = ['AUD', 'CAD', 'CHF', 'EUR', 'GBP', 'JPY', 'NOK', 'NZD', 'SEK']
```

```python
#Simple Portfolio returns using csfs weights
df['simple_portfolio_return'] = returns.mul(portfolio_weights).sum(axis = 1)
```

```python
#lets see both benchmark and simple portfolio return
df[['bench_return', 'simple_portfolio_return' ]]
```

```python
Number_of_days = df.shape[0]
Annualized_bench_return = (((1 + df['bench_return']/100).prod())**(365/Number_of_days) -1)*100
Annualized_portfolio_return = (((1 + df['simple_portfolio_return']/100).prod())**(365/Number_of_days) -1)*100
```

```python
print(Annualized_bench_return)
print(Annualized_portfolio_return)
```

```python
Annualized_benchmark_risk = df['bench_return'].std()*math.sqrt(252)
Annualized_portfolio_risk = df['simple_portfolio_return'].std()*math.sqrt(252)
```

```python
print(Annualized_benchmark_risk)
print(Annualized_portfolio_risk)
```

```python
cum_returns_simple_portfolio = ((1 + df['simple_portfolio_return']/100).cumprod())*100
drawdown_simple_portfolio =  (1 - cum_returns_simple_portfolio.div(cum_returns_simple_portfolio.cummax()))*100
simple_port_max_draw = drawdown_simple_portfolio.max()
```

```python
cum_returns_bench_portfolio = ((1 + df['bench_return']/100).cumprod())*100
drawdown_bench_portfolio =  (1 - cum_returns_bench_portfolio.div(cum_returns_bench_portfolio.cummax()))*100
bench_port_max_draw = drawdown_bench_portfolio.max()
```

```python
table = PrettyTable()
table.align = "l"
table.hrules = 1

table.title = 'Summary Table of Above Calculations'
table.field_names = ['Parameter', 'Value']

table.add_row(['Annualized Benchmark Return %', Annualized_bench_return])
table.add_row(['Annualized Benchmark Risk %', Annualized_benchmark_risk])
table.add_row(['Sharpe Ratio:Benchmark', Annualized_bench_return/Annualized_benchmark_risk])
table.add_row(['Maximum Drawdown:BenchMark Portfolio %', bench_port_max_draw ])
table.add_row(['Annualized Return/Maximum Drawdown:Benchmark Portfolio %', Annualized_bench_return/bench_port_max_draw ])



table.add_row(['Annualized Simple Portfolio Return %', Annualized_portfolio_return])
table.add_row(['Annualized Simple Portfolio Risk %', Annualized_portfolio_risk])
table.add_row(['Sharpe Ratio:Simple Portfolio', Annualized_portfolio_return/Annualized_portfolio_risk])
table.add_row(['Maximum Drawdown:Simple Portfolio %', simple_port_max_draw ])
table.add_row(['Annualized Return/Maximum Drawdown:Simple Portfolio %', Annualized_portfolio_return/simple_port_max_draw ])

print(table)
```

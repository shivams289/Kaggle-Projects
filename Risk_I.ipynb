```python
#For Puts
#Q2(b)
#Reusing the above code
#data: complete dataframe
#option_type: 'CE' or 'PE' as string
def extract_top5_OTM_options(data, option_type):
    q2 = df2.groupby(['SYMBOL', 'EXPIRY_DT'])
    q2_group = q2.apply(lambda x:x[(x.OPTION_TYP == option_type)])
    q2_group.reset_index(inplace=True, drop=True)
    #taking out the underlying's SETTLE_PR
    temp = q2.apply(lambda x:x[(x.OPTION_TYP == 'XX')])
    temp.reset_index(inplace=True, drop=True)
    temp = temp[['SYMBOL', 'EXPIRY_DT', 'SETTLE_PR']]
    temp.rename(columns = {'SETTLE_PR':'SETTLE_PR_UNDERLYING'}, inplace=True)
    #Left Joining underlying SETTLE_PR for a particular SYMBOL, EXPIRY_DT option
    q2 = pd.merge(q2_group, temp, on = ['SYMBOL', 'EXPIRY_DT'], how = 'left')
    #if some NAN's are created after Left join, where we don't have SETTLE_PR
    q2.dropna(axis = 0, inplace =True)
    #filtering out OTM calls
    q2 = q2[q2.STRIKE_PR >= q2.SETTLE_PR_UNDERLYING]
    q2_final = q2[['SYMBOL', 'EXPIRY_DT', 'STRIKE_PR', 'VAL_INLAKH']].groupby(['SYMBOL', 'EXPIRY_DT'], as_index=False).apply(lambda df:df.sort_values('STRIKE_PR', ascending=False).head(5))
    
    return q2_final
    
```

## Question-4

```python
#Question 4
#I am assuming trading_volume here means VAL_INLAKH
q4_day1 = pd.read_csv('fo02AUG2017bhav.csv')
q4_day2 = pd.read_csv('fo03AUG2017bhav.csv')
```

```python
#For each SYMBOL and EXPIRY date we can have multiple OPTION_TYP so we have to take care of that too
#extracting the columns needed for our analysis
q4_day1 = q4_day1[['SYMBOL', 'EXPIRY_DT', 'OPTION_TYP', 'VAL_INLAKH']]
```

```python
q4_day2 = q4_day2[['SYMBOL', 'EXPIRY_DT', 'OPTION_TYP', 'VAL_INLAKH']]
```

```python
final = pd.merge(q4_day1, q4_day2, on = ['SYMBOL', 'EXPIRY_DT', 'OPTION_TYP'], how = 'left', suffixes=['_DAY1','_DAY2'])
```

```python
final
```

```python
#I am taking percentage increase in volumes as my metric as it's not specified if it 
#should be absolute increase or percentage increase over day 2
final['TRADING_VOL_INCREASE'] = ((final.VAL_INLAKH_DAY2 - final.VAL_INLAKH_DAY1)/final.VAL_INLAKH_DAY1)*100
```

```python
final = final[(final.VAL_INLAKH_DAY2 != 0) & (final.VAL_INLAKH_DAY1 !=0)]
```

```python
final.sort_values(by = ['TRADING_VOL_INCREASE'], ascending=False)
```

```python
#Here We are going to check the behavior of distributions in QQ plots
#QQ plots are basically Quantile-Quantile plots of your data with the exact Quantiles of that distribution
#So your data might have 10% of datapoints in 1st Quantile while the actual theoretical distrubtion has 12%
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
import random
import scipy
from scipy.stats import ttest_1samp
import statsmodels.api as sm
import scipy.stats as stats
```

```python
#We take a uniform distribution continuos variable
np_uniform = pd.Series(np.random.uniform(-5, 5, 2000))

#We take normal distribution continuos random variable
np_rand = pd.Series(np.random.normal(-5, 5, 2000))
```

```python
#Distribution plot says that our data is approx. Uniform
sns.displot(np_uniform)
```

```python
#Uniformity is indicated by QQ-lot as all the points are lying on the line
plot1u = sm.qqplot(np_uniform,line='45',fit=True, dist=stats.uniform)
```

```python
#if we fit normal data on a QQ plot based on normal distribution, we can see that Quantiles doesn't lie on our line
plot2u = sm.qqplot(np_uniform,line='45',fit=True, dist=stats.norm)
```

```python
#This is plot w.r.t to a t distribution
plot3u = sm.qqplot(np_uniform,line='45',fit=True, dist=stats.t)
```

```python
#Now we check our random varible distribution plot
sns.displot(np_rand, kde =True)
```

```python
#Here we can see that our normal continuos random varible is approximately normal with a little 
#negative skew indicated by points on the left lying below our 45 degree line
plot1n = sm.qqplot(np_rand,line='45',fit=True, dist=stats.norm)
```

```python
#Now we check if it follows t - distribution
plot2n = sm.qqplot(np_rand,line='45',fit=True, dist=stats.t)
```

```python
#Check for uniform distribution
plot3n = sm.qqplot(np_rand,line='45',fit=True, dist=stats.uniform)
```

```python
#Now lets increase the size of our data or increase 'n' by 100 fold
np_rand2 = pd.Series(np.random.normal(-5, 5, 2000*100))
```

```python
#Now lets see the distribution
sns.displot(np_rand2, kde =True)
```

```python
#Now the QQ-plot
#We can see as we increased the number of data points it became more normal
#Central limit theorem
plot11n = sm.qqplot(np_rand2,line='45',fit=True, dist=stats.norm)
```

```python
plot22n = sm.qqplot(np_rand2,line='45',fit=True, dist=stats.t)
```
